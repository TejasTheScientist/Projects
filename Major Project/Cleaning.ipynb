```python
# import
import pandas as pd
```

```python
data_import = pd.read_csv("./raw_job_data.csv")
data_import.head()
```

```python
data = data_import.copy()
```

```python
# droping 1st column
data = data.drop(columns=data.columns[0], axis = 1)
data.head()
```

```python
data['Job Title'].nunique()
```

```python
# removing (Glassdoor est.) from suffix of salary

d1 = data['Salary Estimate'].str.removesuffix('(Glassdoor est.)')
d1
```

```python
# also removing the $ and k

d1 = d1.str.replace('$','').str.replace('K','')
d1 = d1.str.strip()
d1
```

```python
# checking if something wrong in data
d1.sample(10)
```

```python
# if lenght of salary > 7 we need to handle that data

d1[(d1.str.len() > 7) == True]
```

```python
# removing suffix and prefix

d1 = d1.str.removeprefix('Employer Provided Salary:')
d1 = d1.str.removesuffix('(Employer est.)')
```

```python
# lets see what we get

d1[(d1.str.len() > 7) == True]
```

lets create seperate col for per hour

```python
data['Per hour'] = d1.apply(lambda x : 1 if 'Per Hour' in x else 0)
```

```python
# lets remove the suffix 

d1 = d1.str.removesuffix(' Per Hour')
d1
```

```python
# lets update our column

data['Salary Estimate'] = d1
```

'-1' means information not available

```python
data[data['Salary Estimate'] == '-1'].count()
```

lets drop them

```python
data.drop(index = data[data['Salary Estimate'] == '-1'].index, inplace = True)
```

```python
data['Rating'].unique()
```

-1 rating means not available
lets replace it by mean

```python
data['Rating'].replace(-1.0, data['Rating'].mean(), inplace = True)
```

Lets add more columns in our dataframe

```python
data['Min Salary'] = data['Salary Estimate'].apply(lambda x : int(x.split('-')[0]))
data['Max Salary'] = data['Salary Estimate'].apply(lambda x : int(x.split('-')[1]))
```

lets update min and max salary for per hour

```python
data['Min Salary'] = data.apply(lambda x : x['Min Salary']*2 if x['Per hour'] == 1 else x['Min Salary'] ,axis =1)
data['Max Salary'] = data.apply(lambda x : x['Max Salary']*2 if x['Per hour'] == 1 else x['Max Salary'] ,axis =1)
```

```python
data['Avg Salary'] = (data['Min Salary']+data['Max Salary'])/2
```

Lets also add age of company

```python
import datetime as dt

data['age'] = dt.date.today().year - data.Founded
data.head()
```

lets add columns for skills for job ie python, power bi, aws,excel

```python
data['Python'] = data['Job Description'].apply(lambda x : 1 if 'python' in x.lower() else 0)
data['Power bi'] = data['Job Description'].apply(lambda x : 1 if 'power bi' in x.lower() else 0)
data['Excel'] = data['Job Description'].apply(lambda x : 1 if 'excel' in x.lower() else 0)
data['Aws'] = data['Job Description'].apply(lambda x : 1 if 'aws' in x.lower() else 0)
data.head()
```

lets create new job title

```python
data['Job Title'].unique()
```

```python
def new(title):
    t = title.lower()
    
    if 'data scientist' in t:
        return 'Data Scientist'

    elif 'analyst' in t:
        return 'Data Analyst'

    elif 'manager' in t:
        return 'Manager'

    elif 'machine learning' in t:
        return 'ML Engineer'

    elif 'engineer' in t:
        return 'Data Engineer'

    else:
        return 'na'

data['Title'] = data['Job Title'].apply(new)
```

```python
def position(posi):
    p = posi.lower()

    if ('sr' in p) or ('senior' in p) or ('expert' in p) or ('principal' in p):
        return 'Senior'

    elif ('jr' in p) or ('junior' in p):
        return 'Junior'

    else :
        return 'na'

data['position'] = data['Job Title'].apply(position)
```
