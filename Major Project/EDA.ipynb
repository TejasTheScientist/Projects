```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

```python
import_data = pd.read_csv("./clean_data.csv")
import_data.head()
```

```python
data = import_data.copy()
data.columns
```

```python
data.info()
```

```python
data.describe().T
```

```python
num_cols = ['Rating', 'Min Salary', 'Max Salary',
       'Avg Salary', 'Same place', 'Python', 'Power bi', 'Excel', 'Aws',
       'num_compi']
num_cols
```

```python
c = 0
plt.figure(figsize=(10,25))
for i in num_cols:
    c = c+1
    plt.subplot(5,2,c)
    sns.distplot(data[i])
    plt.title(i)
plt.show()
```

```python
data.corr()
```

```python
corr_cols = ['Rating','Avg Salary', 'Same place', 'Python', 'Power bi', 'Excel', 'Aws','num_compi']
```

```python
plt.figure(figsize = (12,9))
sns.heatmap(data[corr_cols].corr(), annot = True, cmap = 'Blues')
plt.show()
```
